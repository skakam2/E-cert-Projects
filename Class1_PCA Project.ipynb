```python
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from bokeh.plotting import show, figure, output_file
from bokeh.io import show, output_notebook
from bokeh.sampledata.iris import flowers
from sklearn import datasets
from sklearn.decomposition import PCA
%matplotlib inline
```

```python

iris2 = datasets.load_iris()

# Bokeh code
colormap = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
colors = [colormap[x] for x in flowers['species']]

p = figure(title = 'IRIS Dataset: PCA for 2 components in Bokeh Framework')
p.xaxis.axis_label = 'Length 1'
p.yaxis.axis_label = 'Length 2'

X1_reduced = PCA(n_components = 2).fit_transform(iris2.data)
p.circle(X1_reduced[:,0], X1_reduced[:,1], color= colors, fill_alpha=0.2, size=10)

output_notebook(hide_banner=True)
show(p)

#######################################
# 3D display of the IRIS data using PCA
iris2 = datasets.load_iris()
fig = plt.figure(1, figsize = (8,6))
ax = Axes3D(fig, elev = -155, azim = 110)
X_reduced = PCA(n_components = 3).fit_transform(iris2.data)

ax.set_title('IRIS Dataset: PCA for 3 components')
ax.set_xlabel('1st eigenvector')
ax.set_ylabel('2nd eigenvector')
ax.set_zlabel('3rd eigenvector')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.scatter(X_reduced[:,0], X_reduced[:,1], X_reduced[:,2], c=colors, edgecolor='k', s=30)
```
